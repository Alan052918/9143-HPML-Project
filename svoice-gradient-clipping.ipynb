```python
import json
import matplotlib.pyplot as plt
```

```python
clipped = []
with open("svoice-training-results/adam-step/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        clipped.append(d["best"])

# print(clipped)
```

```python
noclip = []
with open("svoice-training-results/adam-step-noclip/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        noclip.append(d["best"])

# print(noclip)
```

```python
plt.plot(clipped, label="Gradient Clipping")
plt.plot(noclip, label="None")
plt.xlabel("Epoch")
plt.ylabel("Valid loss")
plt.legend()
plt.show()
```
