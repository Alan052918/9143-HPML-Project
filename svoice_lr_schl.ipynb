# Validation Loss Results Using Different Learning Rate Schedulers

```python
import json
import matplotlib.pyplot as plt
```

## Adadelta

```python
adadelta_step = []
with open("svoice-training-results/adadelta-step/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        adadelta_step.append(d["best"])

# print(adadelta_step)

adadelta_plateau = []
with open("svoice-training-results/adadelta-plateau/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        adadelta_plateau.append(d["best"])

# print(adadelta_plateau)
```

```python
plt.plot(adadelta_step, label="Adadelta StepLR")
plt.plot(adadelta_plateau, label="Adadelta Plateau")
plt.xlabel("Epoch")
plt.ylabel("Valid loss")
plt.legend()
plt.show()
```
