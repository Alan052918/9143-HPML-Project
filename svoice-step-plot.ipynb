```python
import json
```

```python
adadelta_step = []
with open("svoice-training-results/adadelta-step/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        adadelta_step.append(d["best"])

# print(adadelta_step)
```

```python
adagrad_step = []
with open("svoice-training-results/adagrad-step/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        adagrad_step.append(d["best"])

# print(adagrad_step)
```

```python
adam_step = []
with open("svoice-training-results/adam-step/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        adam_step.append(d["best"])

# print(adam_step)
```

```python
rmsprop_step = []
with open("svoice-training-results/rmsprop-step/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        rmsprop_step.append(d["best"])

# print(rmsprop_step)
```

```python
sgd_step = []
with open("svoice-training-results/sgd-step/history.json", "r") as f:
    data = json.load(f)
    for d in data:
        sgd_step.append(d["best"])

# print(sgd_step)
```

```python
import matplotlib.pyplot as plt

plt.plot(adadelta_step, label="Adadelta")
plt.plot(adagrad_step, label="Adagrad")
plt.plot(adam_step, label="Adam")
plt.plot(rmsprop_step, label="RMSProp")
plt.plot(sgd_step, label="SGD")
plt.xlabel("Epoch")
plt.ylabel("Valid loss")
plt.legend()
plt.show()
```
