```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import re

pattern = "\d+"

opt_list = ['Adadelta', 'Adagrad', 'Adam', 'Nesterov', 'RMSProp', 'SGD']

Adadelta_loss_list = []
Adagrad_loss_list = []
Adam_loss_list = []
Nesterov_loss_list = []
RMSProp_loss_list = []
SGD_loss_list = []

Adadelta_step_time_list = []
Adagrad_step_time_list = []
Adam_step_time_list = []
Nesterov_step_time_list = []
RMSProp_step_time_list = []
SGD_step_time_list = []

def load_data(filename):
    count = 0
    loss_list = []
    step_time_list = []
    with open(filename) as fp:
        Lines = fp.readlines()
        for line in Lines:
            count += 1
            if count %2 !=0 :
                r1 = re.findall(pattern, line.strip())
                loss = r1[0] + '.' + r1[1]
                loss_list.append(float(loss))

                if len(r1) > 4:
                    step_time = r1[3] + '.' + r1[4]
                    step_time_list.append(float(step_time))
    step_time_list.pop(0)
    return loss_list, step_time_list

Adadelta_loss_list, Adadelta_step_time_list = load_data('Adadelta')
Adagrad_loss_list, Adagrad_step_time_list = load_data('Adagrad')
Adam_loss_list, Adam_step_time_list = load_data('Adam')
Nesterov_loss_list, Nesterov_step_time_list = load_data('Nesterov')
RMSProp_loss_list, RMSProp_step_time_list = load_data('RMSProp')
SGD_loss_list, SGD_step_time_list = load_data('SGD')
```

```python
x = range(1, 9)
color_list = ['cyan', 'red', 'orange', 'blue', 'green', 'brown']
f, ax = plt.subplots(1, figsize=(10,10))
plt.plot(x, Adadelta_loss_list[0:8], label="Adadelta", color='cyan')
plt.plot(x, Adagrad_loss_list[0:8], label="Adagrad", color='red')
plt.plot(x, Adam_loss_list[0:8], label="Adam", color='orange')
plt.plot(x, Nesterov_loss_list[0:8], label="Nesterov", color='blue')
plt.plot(x, RMSProp_loss_list[0:8], label="RMSProp", color='green')
plt.plot(x, SGD_loss_list[0:8], label="SGD", color='brown')
plt.legend(fontsize=15)
plt.xlabel('Step')
plt.ylabel('Loss')
plt.show()
```
